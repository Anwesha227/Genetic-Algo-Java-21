```python
#Converting non-numeric columns to numeric columns
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
 

enc.fit(df['Churn'])
df['Churn'] = enc.transform(df['Churn'])




import numpy as np
numeric_only_cols = df.select_dtypes(include=np.number).columns.tolist()
#numeric_only_cols
all_cols = df.columns.tolist()

ob_cols = [x for x in all_cols if x not in numeric_only_cols]
ob_cols
```

```python
ob_cols.remove('HandsetPrice')
ob_cols.remove('CreditRating')
ob_cols.remove('ServiceArea')
```

```python
df = pd.get_dummies(df, columns=['PrizmCode','Occupation'])
```

```python
ob_cols.remove('PrizmCode')
ob_cols.remove('Occupation')
ob_cols.remove('MaritalStatus')

for item in ob_cols:
    print(df[item].unique())
```

```python
for item in ob_cols:
    enc.fit(df[item])
    df[item] = enc.transform(df[item])
```

Convert the 'unknowns' of Handsetprice to 'NaN' values for imputation. Do same for Marital Status


```python
df['HandsetPrice'].replace('Unknown',np.NaN,inplace = True)
df['MaritalStatus'].replace('Yes',1, inplace=True)
df['MaritalStatus'].replace('No',0, inplace=True)
df['MaritalStatus'].replace('Unknown',np.NaN, inplace= True)


for item in df['CreditRating'].unique():
    num = 8-int(item[0])
    df['CreditRating'].replace(item,num, inplace=True)
    
df = df.drop('ServiceArea', axis=1)
```
